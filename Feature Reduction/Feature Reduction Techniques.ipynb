
## Feature Reduction Techniques and Classification Algorithms

## Introduction

 A classification dataset was taken from kaggle. The data was splited into training and testing splits in the ratio of 75-25. And done the Feature scaling,Feature reduction using LDA & SVD techniques and 2 separate train-test datasets were created.
Then we have done the classification using SVM, Naive Bayes and Decision tree Classifier on each of the datasets and analyzed the performance of each Feature reduction technique.

## Dataset Information

Age : Age of the patient

Sex : Sex of the patient

exang: exercise induced angina (1 = yes; 0 = no)

ca: number of major vessels (0-3)

cp : Chest Pain type chest pain type

     Value 1: typical angina
     Value 2: atypical angina
     Value 3: non-anginal pain
     Value 4: asymptomatic
     
trtbps : resting blood pressure (in mm Hg)

chol : cholestoral in mg/dl fetched via BMI sensor

fbs : (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)

rest_ecg : resting electrocardiographic results

     Value 0: normal
     Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
     Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
     
thalach : maximum heart rate achieved

target : 0= less chance of heart attack 1= more chance of heart attack

Source of Dataset : https://www.kaggle.com/rashikrahmanpritom/heart-attack-analysis-prediction-dataset/version/2

## Importing Dataset

```python
import numpy as np
import pandas as pd
import io
import seaborn as sns
```

```python
df=pd.read_csv("heart.csv")
df.head()
```

## Training and Testing Split 

```python
y=df['output']
X=df.drop('output',axis=1)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

# Feature Reduction

## LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda=LDA(n_components=1)
X_train_LDA = lda.fit_transform(X_train, y_train)
X_test_LDA = lda.transform(X_test)
print(X_train_LDA)
```

## SVD

```python
from sklearn.decomposition import TruncatedSVD
svd=TruncatedSVD(n_components=2)
X_train_SVD=svd.fit_transform(X_train, y_train)
X_test_SVD=svd.transform(X_test)
print(X_train_SVD)
```

## SVM Using LDA


```python
from sklearn.svm import SVC
model_svm_lda=SVC(kernel='linear')
model_svm_lda.fit(X_train_LDA, y_train)
y_pred=model_svm_lda.predict(X_test_LDA)

from sklearn.metrics import classification_report, confusion_matrix
print("Confusion Matrix SVM using LDA")
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

## SVM Using SVD


```python
from sklearn.svm import SVC
model_svm_svd=SVC(kernel='linear')
model_svm_svd.fit(X_train_SVD, y_train)
y_pred=model_svm_svd.predict(X_test_SVD)

from sklearn.metrics import classification_report, confusion_matrix
print("Confusion Matrix SVM using SVD")
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

## Naive Bayes Using LDA


```python
from sklearn.naive_bayes import GaussianNB 
model_nb_lda=GaussianNB()
model_nb_lda.fit(X_train_LDA, y_train)
y_pred=model_nb_lda.predict(X_test_LDA)

from sklearn.metrics import classification_report, confusion_matrix
print("Confusion Matrix Naive Bayes Using LDA")
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

## Naive Bayes Using SVD

```python
from sklearn.naive_bayes import GaussianNB 
model_nb_svd=GaussianNB()
model_nb_svd.fit(X_train_SVD, y_train)
y_pred=model_nb_svd.predict(X_test_SVD)

from sklearn.metrics import classification_report, confusion_matrix
print("Confusion Matrix Naive Bayes Using LDA")
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

## Decision Tree Using LDA

```python
from sklearn.tree import DecisionTreeClassifier
model_dt_lda=DecisionTreeClassifier(random_state=np.random)
model_dt_lda.fit(X_train_LDA, y_train)
y_pred=model_dt_lda.predict(X_test_LDA)

from sklearn.metrics import classification_report, confusion_matrix
print("Confusion Matrix Decision Tree Using LDA")
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

## Decision Tree Using SVD

```python
from sklearn.tree import DecisionTreeClassifier
model_dt_svd=DecisionTreeClassifier(random_state=np.random)
model_dt_svd.fit(X_train_SVD, y_train)
y_pred=model_dt_svd.predict(X_test_SVD)

from sklearn.metrics import classification_report, confusion_matrix
print("Confusion Matrix Decision Tree Using SVD")
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

# Inference

Accuracy can be a good measure to evaluate all the algorithhms or models and Feature reduction techniques.

Analysis on Feature Reduction Techniques: LDA performs better than SVD for Naive Bayes and Decision Tree and for SVM SVD performs slightly better i.e. 1% more accurate than LDA. As LDA creates maximize the separation between the 2 classes while SVD tries to make maximum variability between the new features.

Analysis on Classification Algorithms: Performance of all the algorithms (SVM,Naive Bayes & Decision tree) have almost similar accuracy depending upon the feature reductoin technique while SVM using SVD performs slightly better by 1 % for dataset on which LDA is applied.
